## Step 2: Getting the data ready to fit the model

- Add columns from the Incident Date column so we can use Hour, Day of Week, Day of Month, and Month in our Logistic Regression model

```python
# convert to datetime

df['Incident Date'] = pd.to_datetime(df['Incident Date'])
```

```python
df['Hour'] = df['Incident Date'].dt.hour
```

```python
df['Day of Week'] = df['Incident Date'].dt.weekday
```

```python
weekday_map = {0: 'MONDAY', 1: 'TUESDAY', 2: 'WEDNESDAY', 3: 'THURSDAY', 4: 'FRIDAY', 5: 'SATURDAY', 6: 'SUNDAY'}
df['Day of Week'] = df['Day of Week'].map(lambda x: weekday_map[x])
```

```python
df['Month'] = df['Incident Date'].dt.month
```

```python
df['Day of Month'] = df['Incident Date'].dt.day
```

```python
df.head()
```

```python
df.to_csv('v11IncidentPrediction.csv', index=False)
```

```python
df2 = df
```

```python
# Mapping for Overall Race
race_map = {
    'HISPANIC': 0,
    'WHITE': 1,
    'OTHER': 2,
    'BLACK': 3,
    'ASIAN': 4,
    'UNKNOWN': 5,
    'PACIFIC ISLANDER': 6,
    'AMERICAN INDIAN': 7
}
```

```python
df2['Overall Race'] = df['Overall Race'].map(race_map)
```

```python
# Mapping for Day of Week
day_of_week_map = {
    'MONDAY': 0,
    'TUESDAY': 1,
    'WEDNESDAY': 2,
    'THURSDAY': 3,
    'FRIDAY': 4,
    'SATURDAY': 5,
    'SUNDAY': 6
}
```

```python
df2['Day of Week'] = df['Day of Week'].map(day_of_week_map)
```

```python
df2.to_csv('v11NumericIncidentPrediction.csv', index=False)
```

## Step 3: Fitting the model: Decision Tree

```python
# Define the feature columns and target column
feature_columns = ["Victim Age", "Overall Race", "Zip Code", "Domestic Violence Incident", "Hour", "Day of Week", "Day of Month", "Month"]
target_column = "CIBRS Offense Description"
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
X, y = df[feature_columns], df[target_column]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(X, y)
```

```python
# Function to make predictions
def predict_incident(victim_age, overall_race, zip_code, domestic_violence_incident, hour, day_of_week, day_of_month, month):
    # Map the input values to the corresponding numerical values
    overall_race = race_map[overall_race]
    day_of_week = day_of_week_map[day_of_week]
    
    # Create a DataFrame for the input values
    input_data = pd.DataFrame({
        'Victim Age': [victim_age],
        'Overall Race': [overall_race],
        'Zip Code': [zip_code],
        'Domestic Violence Incident': [domestic_violence_incident],
        'Hour': [hour],
        'Day of Week': [day_of_week],
        'Day of Month': [day_of_month],
        'Month': [month]
    })
    
    # Make a prediction
    prediction = clf.predict(input_data)
    return prediction[0]
```

```python
# Example usage
predicted_incident = predict_incident(25, 'WHITE', 92101, 0, 14, 'MONDAY', 15, 6)
print(f"Predicted Incident: {predicted_incident}")
```
